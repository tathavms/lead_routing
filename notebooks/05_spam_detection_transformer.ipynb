```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoConfig
import torch
```

```python

tokenizer = AutoTokenizer.from_pretrained("../models/spam_detection_model")
model = AutoModelForSequenceClassification.from_pretrained("../models/spam_detection_model")
model.eval()
config = AutoConfig.from_pretrained("../models/spam_detection_model")
```

```python
text = "Immediate Help Needed: Technical Problem with Cloud SaaS Service..."

inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
inputs.pop('token_type_ids', None)
```

```python
with torch.no_grad():
    outputs = model(**inputs)

probabilities = torch.softmax(outputs.logits, dim=1)
pred_idx   = torch.argmax(probabilities, dim=1)[0].item()
confidence = probabilities[0][pred_idx].item()

print(f"Prediction: {config.id2label[pred_idx]} ({confidence:.2%})")
```
